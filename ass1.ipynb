```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load data
df = pd.read_csv(r'C:\Users\rbhoo\Downloads\archive\personality_dataset.csv')  # Replace with your actual file name

# Step 2: Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Now your columns are like:
# ['time_spent_alone', 'stage_fear', 'social_event_attendance', 'going_outside',
#  'drained_after_socializing', 'friends_circle_size', 'post_frequency', 'personality']

# Step 3: Check data and handle missing values if needed
print(df.info())
df = df.dropna()

# Step 4: Set personality column as categorical
df['personality'] = df['personality'].astype('category')
```

```python
sns.countplot(data=df, x='personality')
plt.title('Extroverts vs. Introverts')
plt.xlabel('Personality Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(data=df, x='personality', y='time_spent_alone')
plt.title('Average Time Spent Alone by Personality')
plt.xlabel('Personality Type')
plt.ylabel('Time Spent Alone')
plt.show()
```

```python
sns.boxplot(data=df, x='personality', y='friends_circle_size')
plt.title('Friends Circle Size by Personality Type')
plt.xlabel('Personality Type')
plt.ylabel('Friends Circle Size')
plt.show()
```
